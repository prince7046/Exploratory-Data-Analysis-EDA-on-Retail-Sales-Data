```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x='Category'
)

plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Calories'],
    bins=20,
    kde=True
)

plt.show()
```

```python
top10 = df.nlargest(
    10,
    'Calories'
)

top10[['Item','Calories']]
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=top10,
    x='Calories',
    y='Item'
)

plt.show()
```

```python
low10 = df.nsmallest(
    10,
    'Calories'
)

low10[['Item','Calories']]
```

```python
avg_cal = df.groupby(
    'Category'
)['Calories'].mean()

avg_cal.sort_values().plot(
    kind='bar',
    figsize=(10,5)
)

plt.ylabel("Average Calories")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x='Total Fat',
    y='Calories'
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x='Sugars',
    y='Calories'
)

plt.show()
```

```python
plt.figure(figsize=(14,10))

corr = df.corr(
    numeric_only=True
)

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.show()
```

```python
protein = df.groupby(
    'Category'
)['Protein'].mean()

protein.plot(
    kind='bar',
    figsize=(10,5)
)

plt.ylabel("Average Protein")
plt.show()
```
